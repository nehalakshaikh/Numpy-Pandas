```python
def run_scored(batsman_name):
    vk=delivery[delivery['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
# it give highest run score against any team bye given player    
```

```python
# find the most destructive death over batsman in the history of ipl
# strike rate=(no. of runs/no. of balls)*100
# min batsman 200 balls in over 16-20
```

```python
death_over=delivery[delivery['over']>15]
```

```python
all_batsman=death_over.groupby('batsman')['ball'].count()
```

```python
x=all_batsman>200
all_batsman[x].shape
```

```python
batsman_list=all_batsman[x].index.tolist() 
#tolist convert index into list 
```

```python
# runs score by all these 43 batsman
# balls played by these 43 batsman
```

```python
final=death_over[death_over['batsman'].isin(batsman_list)]
```

```python
run=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
sr=(run/balls)*100
```

```python
sr.sort_values(ascending=False).head()
```

# pivot_table

```python
# in pivot table we have to give 4 thing first is index(row), second is columns,
# thirst is values= what data will present in dataframe, forth is aggregate func(eg-mean,min,max,count) 
```

```python
mask=delivery['batsman_runs']==6
sixes=delivery[mask]
sixes.shape
```

```python
pt=sixes.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')
pt
```
